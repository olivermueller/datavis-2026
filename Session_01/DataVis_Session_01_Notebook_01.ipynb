## Your First Plot

### The Figure/Axes Architecture

Matplotlib uses a two-level hierarchy that is important to understand:

- **`Figure`** — the outermost container, like a canvas or a blank sheet of paper. It holds everything and controls the overall size of the image.
- **`Axes`** — a single plot within the figure. It contains the actual chart elements: data, axes lines, ticks, labels, and title.

    ```
    Figure
    └── Axes (one or more)
        ├── x-axis, y-axis
        ├── data (scatter, line, bar, ...)
        ├── title, labels
        └── legend
    ```

When you call `fig, ax = plt.subplots()`, you create both at once. A figure can contain *multiple* axes (e.g., in small multiples), but each axes belongs to exactly one figure. Most customization happens on the `ax` object.

````python
fig, ax = plt.subplots(figsize=(10, 6))
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(gapminder["gdpPercap"], gapminder["lifeExp"], marker="o", linestyle="none")
plt.show()
````

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(gapminder["gdpPercap"], gapminder["lifeExp"])
plt.show()
````

## Mapping Data to Visuals

````python
colors = {'Africa':'tab:brown', 'Asia':'tab:red', 'Europe':'tab:grey', 'Americas':'tab:green', 'Oceania':'tab:blue'}

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(gapminder["gdpPercap"], gapminder["lifeExp"], s=minmax_scale(gapminder["pop"], (10,1000)), c=gapminder["continent"].map(colors), alpha=0.3)

# the next lines are a trick to create a legend, we we use Seaborn or Plotly later in this course this is done automatically 
for continent, color in colors.items():
    ax.scatter([], [], color=color, label=continent)
ax.legend()

plt.show()
````

## Adding Layers

````python
b1, b0 = np.polyfit(gapminder["gdpPercap"], gapminder["lifeExp"], deg=1)

fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(gapminder["gdpPercap"], gapminder["lifeExp"])
ax.plot(gapminder["gdpPercap"], b0 + b1*gapminder["gdpPercap"], color="red")
plt.show()
````

## Tuning the Plot

````python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(gapminder["gdpPercap"], gapminder["lifeExp"])
ax.set_title("GDP per capita vs. Life expectancy")
ax.set_xlabel("GDP per capita")
ax.set_ylabel("Life expectancy")
ax.set_xlim(0, 50000)
ax.set_ylim(0, 100)
ax.set_xticks(np.arange(0, 50000, 10000))
ax.set_yticks(np.arange(0, 100, 10))
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${x:,.0f}"))
plt.show()
````

## Small Multiples

````python
gapminder_europe = gapminder[gapminder["continent"] == "Europe"]
gapminder_asia = gapminder[gapminder["continent"] == "Asia"]
````

````python
fig, ax = plt.subplots(2, 1, figsize=(10, 6), sharex=True, sharey=True)
ax[0].scatter(gapminder_europe["gdpPercap"], gapminder_europe["lifeExp"])
ax[1].scatter(gapminder_asia["gdpPercap"], gapminder_asia["lifeExp"])
ax[0].set_title("Europe")
ax[1].set_title("Asia")
for i in range(2):
    ax[i].set_ylabel("Life expectancy")
    ax[i].set_xscale("log")
    ax[i].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${x:,.0f}"))
ax[1].set_xlabel("GDP per capita")
plt.show()
````

## Theming

Matplotlib has a number of built-in style sheets that can be used to change the overall appearance of your plots (https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html). You can also create your own custom themes by defining a set of style parameters.

````python
with plt.style.context('fivethirtyeight'):
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.scatter(gapminder["gdpPercap"], gapminder["lifeExp"])
    ax.set_title("GDP per capita vs. Life expectancy")
    ax.set_xlabel("GDP per capita")
    ax.set_ylabel("Life expectancy")
    ax.set_xlim(0, 50000)
    ax.set_ylim(0, 100)
    ax.set_xticks(np.arange(0, 50000, 10000))
    ax.set_yticks(np.arange(0, 100, 10))
    ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, pos: f"${x:,.0f}"))
    plt.show()
````

## Summary

In this session, you learned the core building blocks of data visualization with Matplotlib:

| Concept | What you did |
|---|---|
| **Figure / Axes architecture** | `fig, ax = plt.subplots()` creates a canvas and a plot area |
| **Visual encodings** | Mapped data variables to position (x/y), color, and size |
| **Layers** | Added a regression trend line on top of a scatter plot |
| **Axis customization** | Set titles, labels, tick marks, and number formatters |
| **Scales** | Switched to a logarithmic x-axis to reveal hidden structure |
| **Small multiples** | Used `plt.subplots(n, m)` to compare subsets side by side |
| **Theming** | Applied a built-in style sheet without affecting other plots |

**Key principle:** Every visual element in a chart is a mapping from a data variable to a visual channel (position, color, size, shape). Choosing the right mapping is the core skill of data visualization.
